```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
city_path = '/kaggle/input/indian-5000-cities-weather-data/w_d_2/Jaipur_2.csv'
df = pd.read_csv(city_path, parse_dates=['date'])
```

```python
df.isna().sum()
```

```python
df.set_index('date')
```

```python
df.drop(columns=['wind_direction_10m', 'wind_direction_100m'], inplace=True)
df.head()
```

```python
df.drop(columns=['Unnamed: 0'], inplace=True)
df.head()
```

```python
df_c = df.set_index('date')
```

```python
df_c.head()
```

```python
agg_logic = {
    'temperature_2m' : 'mean',
    'relative_humidity_2m' : 'mean', 
    'dew_point_2m' : 'mean',
    'apparent_temperature' : 'mean',
    'precipitation' : 'sum',
    'rain' : 'sum',
    'snowfall' : 'sum',
    'snow_depth' : 'mean',
    'pressure_msl' : 'mean',
    'surface_pressure' : 'mean',
    'cloud_cover' : 'mean',
    'cloud_cover_low' : 'mean',
    'cloud_cover_mid' : 'mean',
    'cloud_cover_high' : 'mean',
    'wind_speed_10m' : 'max',
    'wind_speed_100m' : 'max', 
    'wind_gusts_10m' : 'max'
}
```

```python
df_daily = df_c.resample('D').agg(agg_logic)
df_daily = df_daily.round(2)
df_daily.head()
```

```python
df_daily.isna().sum()
```

```python
df_daily['month'] = df_daily.index.month
df_daily['day_of_year'] = df_daily.index.dayofyear
df_daily['day_of_week'] = df_daily.index.dayofweek 
df_daily['year'] = df_daily.index.year
```

```python
df_daily['temp_lag_1'] = df_daily['temperature_2m'].shift(1)


df_daily['temp_lag_2'] = df_daily['temperature_2m'].shift(2) 
df_daily['precip_lag_1'] = df_daily['precipitation'].shift(1) 
```

```python
df_daily['temp_rolling_mean_7'] = df_daily['temperature_2m'].shift(1).rolling(window=7).mean()
```

```python
df_daily.head()
```

```python
df_daily.tail()
```

```python
df_daily.isna().sum()
```

```python
df_final = df_daily.dropna()
df_final.head()
```

```python
df_final.isna().sum()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='temp_lag_1', y='temperature_2m', data=df_final, alpha=0.5)
plt.title("Today's Temperature vs. Yesterday's Temperature", fontsize=16)
plt.xlabel("Yesterday's Avg Temperature (°C)", fontsize=12)
plt.ylabel("Today's Avg Temperature (°C)", fontsize=12)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='temp_rolling_mean_7', y='temperature_2m', data=df_final, alpha=0.5)
plt.title("Today's Temperature vs. Last 7-Day Average Temperature", fontsize=16)
plt.xlabel("7-Day Rolling Avg Temperature (°C)", fontsize=12)
plt.ylabel("Today's Avg Temperature (°C)", fontsize=12)
plt.grid(True)
plt.show()
```

```python
columns_for_heatmap = [
    'temperature_2m',         
    'apparent_temperature',
    'relative_humidity_2m',
    'precipitation',
    'wind_speed_10m',
    'cloud_cover',
    'month',                  
    'day_of_year',
    'temp_lag_1',
    'temp_rolling_mean_7'
]

correlation_matrix = df_final[columns_for_heatmap].corr()


plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Key Weather Features', fontsize=16)
plt.show()
```

```python
y = df_final['temperature_2m']
X = df_final.drop(columns = ['temperature_2m'])
y.head()
X.head()
```

```python
y.head()
```

```python
X.head()
```

```python
split_point = int(len(X)*0.8)
X_test, X_train = X[split_point:], X[:split_point]
y_train, y_test = y[:split_point], y[split_point:]
print(f"X_train: {X_train.shape}")
print(f"y_train: {y_train.shape}")
```

```python
import xgboost as xgb
from sklearn.metrics import mean_absolute_error
```

```python
model = xgb.XGBRegressor(n_estimators = 1000, early_stopping_rounds = 50, learning_rates = 0.01)
```

```python
model.fit(X_train, y_train, eval_set=[(X_test, y_test)], verbose=100)
```

```python
predictions = model.predict(X_test)
mae = mean_absolute_error(y_test, predictions)
print(f"MAE{mae:.2f}")
```

```python
results = pd.DataFrame({'Actual': y_test, 'Predicted': predictions})

plt.figure(figsize=(15, 7))
results['Actual'].plot(label='Actual Temperature')
results['Predicted'].plot(label='Predicted Temperature', alpha=0.8)
plt.title('Actual vs. Predicted Temperature on Test Data', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Temperature (°C)', fontsize=12)
plt.legend()
plt.grid(True)
plt.show()
```

```python
import joblib

model.save_model('weather_prediction_model_1.json') # <-- CHANGE THIS LINE

model_columns = X.columns
joblib.dump(model_columns, 'weather_model_columns_1.joblib')

print("Model and columns have been saved successfully!")
```
